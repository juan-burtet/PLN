```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url_interesse = 'https://institucional.ufpel.edu.br/servidores/id/36186'

html = urlopen(url_interesse)

soup = BeautifulSoup(html.read())
```

```python
import re

linksProjetos = soup.find_all('a', href=re.compile("/projetos/id/"))

for projeto in linksProjetos:
  print(projeto)
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('universal_tagset')

def get_nouns(input_text):
  from nltk.tokenize import word_tokenize

  text = word_tokenize(input_text)
  doc = nltk.pos_tag(text, tagset='universal')

  list_of_nouns = ['NOUN']
  result= []
  for token in doc:
    if (token[1] in list_of_nouns):
      result.append(token[0])
  return result
```

```python
import urllib.parse

path = 'https://institucional.ufpel.edu.br'

all_nouns = []

for link in linksProjetos:
  
  print(f'PROJETO: {link.getText()}')
  
  html_projeto = urlopen(urllib.parse.urljoin(path,link.get('href')))

  soup_resumo = BeautifulSoup(html_projeto.read())

  resumo = soup_resumo.find('div', text = re.compile('Resumo'))
  
  texto_resumo = resumo.findNext('div').contents[0] # indice para pegar o primeiro elemento do conteúdo do div, que é o proprio texto do resumo

  nouns = get_nouns(texto_resumo)
  
  print(texto_resumo)

  all_nouns.append(nouns)
```

```python
all_nouns = [set(l_) for l_ in all_nouns] # remove os substantivos que aparecerem mais de uma vez em um mesmo resumo de projeto

#set.intersection(all_nouns[7],all_nouns[8]) 

set.intersection(*all_nouns) 
```
